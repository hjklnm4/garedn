```python
import matplotlib.pyplot as plt
import numpy as np
# 3차원 그래프를 그리기 위해서 from mpl_toolkits.mplot3d import Axes3D를 추가해줍니다.
from mpl_toolkits.mplot3d import Axes3D
import math


n = 100

cmin, cmax = 0, 2

#사용자의 갑 입력
x1 = 2000 #가격
y1 = 30 #인쇄속도
z1 = 3800 #인쇄해상도

#보유 제품의 값입력
x = [2600,2800]
y = [20,15]
z = [2400,4600]


plt.rcParams["figure.figsize"] = (7, 7)
fig = plt.figure()


ax = fig.add_subplot(111, projection='3d')

#사용자의 값 그래프에 표시
ax.scatter(x1,y1,z1,  marker='o', s=15, c='blue')
#보유제품의 값 그래프에 표시
ax.scatter(x,y,z, marker='o', s=15, c='green')


plt.show()
```

```python
#점과 점사이의 거리를 구하는 공식
def distance(x1, y1, z1, x2, y2, z2):
    result = math.sqrt( math.pow(x1 - x2, 2) + math.pow(y1 - y2, 2) + math.pow(z1 - z2, 2))
    return result
```

```python
result = distance(x[0], y[0], z[0], x1, y1, z1)
print(result)
```
