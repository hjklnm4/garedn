```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 3차원 그래프를 그리기 위해서 from mpl_toolkits.mplot3d import Axes3D를 추가해줍니다.
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
data = pd.read_excel('./printer_data_new2.xlsx')
```

```python
data.head()
```

```python


cmin, cmax = 0, 7


x=[]
y=[]
z=[]

x1=2125
y1=15
z1=3800


color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(5092)])

for i in range(0,5092):
    x.append(data.loc[i]['가격'] / 1000)
    y.append(data.loc[i]['인쇄 속도'])
    z.append(data.loc[i]['인쇄 해상도'])
    
plt.rcParams["figure.figsize"] = (15, 15)
fig = plt.figure()


ax = fig.add_subplot(111, projection='3d')

#사용자의 값 그래프에 표시
ax.scatter(x1,y1,z1,  marker='o', s=15, c='blue')
#보유제품의 값 그래프에 표시
ax.scatter(x, y, z, marker='o', s=15, c=color, cmap='Oranges')


plt.show()
```

```python
#점과 점사이의 거리를 구하는 공식
def distance(x1, y1, z1, x2, y2, z2):
    result = math.sqrt( math.pow(x1 - x2, 2) + math.pow(y1 - y2, 2) + math.pow(z1 - z2, 2))
    return result
```

```python
result=[]
for i in range(0,5902):
    a = distance(x[i], y[i], z[i], x1, y1, z1)
    result.append(a)
```

```python
idx=result.index(min(result))
print(data['상품명'][idx])
```
